```python
class Canvas:
    def __init__(self, width, height):
        self.width = width
        self.height = height
        # Empty canvas is a matrix with element being the "space" character
        self.data = [[' '] * width for i in range(height)]

    def set_pixel(self, row, col, char='*'):
        self.data[row][col] = char

    def get_pixel(self, row, col):
        return self.data[row][col]
    
    def clear_canvas(self):
        self.data = [[' '] * self.width for i in range(self.height)]
    
    def v_line(self, x, y, w, **kargs):
        for i in range(x,x+w):
            self.set_pixel(i,y, **kargs)

    def h_line(self, x, y, h, **kargs):
        for i in range(y,y+h):
            self.set_pixel(x,i, **kargs)
            
    def line(self, x1, y1, x2, y2, **kargs):
        slope = (y2-y1) / (x2-x1)
        for y in range(y1,y2):
            x= int(slope * y)
            self.set_pixel(x,y, **kargs)
            
    def display(self):
        print("\n".join(["".join(row) for row in self.data]))

class Shape:
    def __init__(self, area, perimeter, x=0, y=0):
        self.__area=area
        self.__perimeter=perimeter
        self.__x=x
        self.__y=y

    def area(self):
        #Calculate the area
        raise NonImplementedError("Subclass must use area function")


    def perimeter(self):
        #Calcualte perimeter
        raise NonImplementedError("Subclass must use perimeter function""")


    def get_x_coord(self):
        #Get the x coordinate for the origin of the shape
        raise NonImplementedError("Use get_x_coord function for this subclass")


    def get_y_coord(self):
        #get the y coordinate for the origin of the shape
        raise NonImplementedError("Use the get_y_coord function for this subclass")
        
    def overlaps(self, other):
        #Check if this shape overlaps with another shape.
        raise NotImplementedError("use overlaps function for this subclass")

    def __repr__(self):
        return f"{self.__class__.__name__}({self.__dict__})"

class Triangle(Shape):
    def __init__(self, height, base, side_a, side_b, x=0, y=0):
        self.__height=height
        self.__base=base
        self.__side_a=side_a
        self.__side_b=side_b
        self.__x=x
        self.__y=y

    def area(self):
        area=(self.__height*self.__base)/2
        return area

    def perimeter(self):
        perimeter=self.__base+self.__side_a+self.__side_b
        return perimeter

    def get_x_coord(self):
        return self.__x

    def get_y_coord(self):
        return self.__y

    def get_points_perimeter(self, n):
        point_list=[]

        #define the vertices
        vertice_1 = [(self.__x, self.__y)]
        vertice_2 = [(self.__x + self.__base, self.__y)]
        vertice_3 = [(self.__x, self.__y + self.__height)]

        #define steps
        side_lengths = [self.__base, self.__side_a, self.__side_b]
        total_perimeter = sum(side_lengths)
        step = total_perimeter / n

        for i in range(n):
            t = i / n
            if t <= 1/3:
                # Along base
                x = self.__x + t * self.__base
                y = self.__y
            elif t <= 2/3:
                # Along height side
                t_new = (t - 1/3) * 3
                x = self.__x
                y = self.__y + t_new * self.__height
            else:
                # Along third side
                t_new = (t - 2/3) * 3
                x = self.__x + t_new * (self.__x - (self.__x + self.__base))
                y = self.__y + t_new * (self.__y + self.__height)
            point_list.append((x, y))

        return point_list

    def overlaps(self, other):
        #test to see if the shapes overlap, True if it does
        # Check for intersection of any two sides
        if self.intersects_line(other.get_vertex1(), other.get_vertex2()) or \
           self.intersects_line(other.get_vertex2(), other.get_vertex3()) or \
           self.intersects_line(other.get_vertex3(), other.get_vertex1()) or \
           other.intersects_line(self.get_vertex1(), self.get_vertex2()) or \
           other.intersects_line(self.get_vertex2(), self.get_vertex3()) or \
           other.intersects_line(self.get_vertex3(), self.get_vertex1()):
            return True

        # Check if one triangle is completely inside the other
        if self.contains_point(other.get_vertex1()) and \
           self.contains_point(other.get_vertex2()) and \
           self.contains_point(other.get_vertex3()):
            return True
        if other.contains_point(self.get_vertex1()) and \
           other.contains_point(self.get_vertex2()) and \
           other.contains_point(self.get_vertex3()):
            return True

        return False

    def intersects_line(self, p1, p2):
        #Tests if a line segment intersects this triangle. True if intersects
        v1, v2, v3 = self.__vertex1, self.__vertex2, self.__vertex3

        # Check if the line segment intersects any of the triangle's sides
        for i in range(3):
            v0 = v1 if i == 0 else v2 if i == 1 else v3
            v1 = v2 if i == 0 else v3 if i == 1 else v1
            v2 = v3 if i == 0 else v1 if i == 1 else v2

            if self.ccw(v0, v1, p1) != self.ccw(v0, v1, p2) and \
               self.ccw(v0, v2, p1) != self.ccw(v0, v2, p2):
                return True

        return False

    def ccw(p1, p2, p3):
        #Check if the points are couter-clockwise
        #1 if the points are counter-clockwise, -1 if clockwise, 0 if collinear.
        return (p2[1] - p1[1]) * (p3[0] - p1[0]) - (p2[0] - p1[0]) * (p3[1] - p1[1])

        
    def contains_point(self, x, y):
        # Calculate the area of the original triangle
        A = self.area()

        # Helper function to compute the area of a triangle given three vertices
        def area_of_triangle(x1, y1, x2, y2, x3, y3):
            return abs((x1 * (y2 - y3) + x2 * (y3 - y1) + x3 * (y1 - y2)) / 2.0)

        # Vertices of the triangle
        v1 = (self.__x, self.__y)
        v2 = (self.__x + self.__base, self.__y)
        v3 = (self.__x, self.__y + self.__height)

        # Calculate the areas with respect to the point and the triangle vertices
        A1 = area_of_triangle(x, y, v2[0], v2[1], v3[0], v3[1])
        A2 = area_of_triangle(v1[0], v1[1], x, y, v3[0], v3[1])
        A3 = area_of_triangle(v1[0], v1[1], v2[0], v2[1], x, y)

        # Check if the sum of A1, A2, and A3 is equal to A within a certain deviation)
        return abs(A - (A1 + A2 + A3)) < 1e-6

    def draw(self, canvas):
        points = self.get_points_perimeter(10)  # Adjust the number of points as needed

        # Draw the triangle using lines
        canvas.line(points[0][0], points[0][1], points[1][0], points[1][1])
        canvas.line(points[1][0], points[1][1], points[2][0], points[2][1])
        canvas.line(points[2][0], points[2][1], points[0][0], points[0][1])

    def __repr__(self):
        return f"Triangle({self.__height}, {self.__base}, {self.__side_a}, {self.__side_b}, {self.__x}, {self.__y})"
    
class Rectangle(Shape):
    def __init__(self, length, width, x=0, y=0):
        self.__width=width
        self.__length=length
        self.__x=x
        self.__y=y

    def area(self):
        area=self.__width*self.__length
        return area

    def perimeter(self):
        perimeter=(self.__width*2)+(self.__length*2)
        return perimeter

    def access_width(self):
        return self.__width

    def access_length(self):
        return self.__length

    def get_x_coord(self):
        return self.__x

    def get_y_coord(self):
        return self.__y

    def overlaps(self, other):
        #Test if shapes overlap, returns True if it does
        x1, y1 = self.get_x(), self.get_y()
        x2 = x1 + self.get_length()
        y2 = y1 + self.get_width()

        x3, y3 = other.get_x(), other.get_y()
        x4 = x3 + other.get_length()
        y4 = y3 + other.get_width()

        return not (x2 < x3 or x1 > x4 or y2 < y3 or y1 > y4)

    def draw(self, canvas):
        x1, y1 = self.get_x_coord(), self.get_y_coord()
        width, height = self.access_length(), self.access_width()

        # Draw the rectangle on the canvas
        canvas.h_line(x1, y1, height)
        canvas.v_line(x1, y1, width)
        canvas.h_line(x1 + width, y1, height)
        canvas.v_line(x1 + width, y1 + height, -width)

    def __repr__(self):
        return f"Rectangle({self.__length}, {self.__width}, {self.__x}, {self.__y})"

class Circle(Shape):
    def __init__(self, radius, x=0, y=0):
        self.__radius=radius
        self.__x=x
        self.__y=y

    def area(self, pi=3.14):
        area=pi*self.__radius**2
        return area

    def perimeter(self, pi=3.14):
        perimeter=2*pi*self.__radius
        return perimeter

    def get_radius(self):
        return self.__radius

    def get_x_coord(self):
        return self.__x

    def get_y_coord(self):
        return self.__y

    def overlaps(self, other):
        #Tests if this circle overlaps, returns True if it does
        distance_between_centers = self.distance_to(other)
        sum_of_radii = self.get_radius() + other.get_radius()
        return distance_between_centers <= sum_of_radii

    def distance_to(self, other):
        #Calculates the distance between the centers of this circle and another circle
        x1, y1 = self.get_center()
        x2, y2 = other.get_center()
        return ((x2 - x1) ** 2 + (y2 - y1) ** 2) ** 0.5

    def draw(self, canvas):
        center_x, center_y = self.get_x_coord(), self.get_y_coord()
        radius = self.get_radius()

    def __repr__(self):
        return f"Circle({self.__radius}, {self.__x}, {self.__y})"

class CompoundShape(Shape):
    def __init__(self, shapes=[]):
        #Remember the list of shapes
        self.shapes = shapes

    def add_shape(self, shape):
        #append the shapes into the list
        self.shapes.append(shape)

    def area(self):
        #Calculate the area of the shapes
        total_area = 0
        for shape in self.shapes:
            total_area += shape.area()
        return total_area

    def perimeter(self):
        #Calculate the perimeter of the shapes
        total_perimeter = 0
        for shape in self.shapes:
            total_perimeter += shape.perimeter()
        return total_perimeter

    def get_x_coord(self):
        #Return the minimum x-coordinate of all shapes within the compound shape
        min_x = float('inf')
        for shape in self.shapes:
            min_x = min(min_x, shape.get_x_coord())
        return min_x

    def get_y_coord(self):
        #Returns the minimum y-coordinate of all shapes within the compound shape.
        min_y = float('inf')
        for shape in self.shapes:
            min_y = min(min_y, shape.get_y_coord())
        return min_y

    def overlaps(self, other):
        #Check if the shape overlaps, returns True if there's overlap
        for shape in self.shapes:
            if shape.overlaps(other):
                return True
        return False

    def draw(self, canvas):
        #Draw on the canvas
        for shape in self.shapes:
            if isinstance(shape, Triangle):
                # Extract points and draw triangle using canvas functions
                points = shape.get_points_perimeter(10)
                canvas.v_line(points[0][0], points[0][1], 1)
                canvas.line(points[0][0], points[0][1], points[1][0], points[1][1])
                canvas.line(points[1][0], points[1][1], points[2][0], points[2][1])
            elif isinstance(shape, Rectangle):
                # Extract coordinates and draw rectangle using canvas functions
                x1, y1 = shape.get_x_coord(), shape.get_y_coord()
                width, height = shape.access_length(), shape.access_width()
                canvas.h_line(x1, y1, height)
                canvas.v_line(x1, y1, width)
                canvas.h_line(x1 + width, y1, height)
                canvas.v_line(x1 + width, y1 + height, -width)
            elif isinstance(shape, Circle):
                # Extract center and radius and draw circle using canvas functions
                center_x, center_y = shape.get_x_coord(), shape.get_y_coord()
                radius = shape.get_radius()

class RasterDrawing:
    def __init__(self, width, height):
        self.canvas = Canvas(width, height)
        self.shapes = [] #Remeber what shapes are drawn

    #Function to add shapes ot the canvas
    def add_shape(self, shape):
        self.shapes.append(shape)

    #Visualize the information added onto the canvas
    def paint(self):
        self.canvas.clear_canvas()
        for shape in self.shapes:
            shape.draw(self.canvas)
    
    #Modify the shape on the canvas
    def modify_shape(self, index, new_shape):
        self.shapes[index] = new_shape

    #Show the shapes
    def display(self):
        self.canvas.display()

    #Save the file
    def save(self,filename):
        f=open(filename,"w")
        f.write(self.__repr__())
        f.close()

def foo_loader(filename):
    f=open(filename,"r")
    tmp=eval(f.read())
    f.close()
    return tmp
"""def save(self, filename):
        with open(filename, 'w') as f:
            for shape in self.shapes:
                f.write(repr(shape) + '\n')

    #load the file
    def load(self, filename):
        shapes = []
        with open(filename, 'r') as f:
            for line in f:
                shape = eval(line)
                shapes.append(shape)
        self.shapes = shapes
        return self"""
```
